```python
import requests
from bs4 import BeautifulSoup

def get_soup(url):
    response = requests.get(url)
    return BeautifulSoup(response.content, 'html.parser') if response.status_code == 200 else None

def extract_links(soup):
    link_queue = []
    for h4_tag in soup.find_all('h4'):
        link_element = h4_tag.find('a')
        if link_element:
            name = h4_tag.text.strip()
            link = 'https://webscraper.io' + link_element.get('href')
            link_queue.append((name, link))
    return link_queue

def extract_swatches(url, product_name):
    soup = get_soup(url)
    if soup:
        swatch_items = soup.find_all('div', class_='product-wrapper card thumbnail')
        for item in swatch_items:
            swatches = item.find('div', class_='swatches')
            if swatches:
                swatch_values = [button['value'] for button in swatches.find_all('button', class_='btn swatch') if 'disabled' not in button.get('class', [])]
                primary_value = swatches.find('button', class_='btn swatch btn-primary active')['value']
                swatch_values.append(primary_value)

                price = item.find('h4', class_='price').text.strip()

                print(f"Product: {product_name} - Price: {price} - Storage variants: {', '.join(swatch_values)}")

url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
soup = get_soup(url)

if soup:
    link_queue = extract_links(soup)
    for name, link in link_queue:
        print(f"Processing product: {name}")
        extract_swatches(link, name)
else:
    print(f"Failed to retrieve the webpage.")
```
